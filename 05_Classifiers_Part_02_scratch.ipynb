## 1. Overview of Decision Tree Classifiers
Key components involved in building a decision tree classifier include:

- **Question**: Represents a query about a feature (e.g., "Is Color == Green?"). You'll need to define how to match feature values against these questions.
- **Nodes**:
  - ***Decision Node***: Asks a question and branches based on the answer. You will create a structure to handle these nodes.
  - ***Leaf Node***: Contains the final classification or prediction based on the data that reaches this node. This will need a mechanism to store class counts.

## 4. **Reminder**: Functions vs. Classes

###**Function**

**What it is**: A block of reusable code that performs a specific task. It takes ***inputs*** (called parameters), processes them, and usually ***returns an output***.

**Purpose**: Help you organize your code into manageable pieces. You can call on a function whenever you need to perform that task without rewriting the same code.

**Example**:

```python
def add_numbers(a, b):
    return a + b

result = add_numbers(3, 5)
print(result)
```
###**Class**

**What it is**: A class is a blueprint for creating objects. It can bundle data (attributes) and functions (methods) that operate on that data together in one place.

**Purpose**: Classes are used to model entities or concepts. They allow you to create multiple objects with similar properties and behaviors without repeating code.

**Example**:
```python
class Dog:
    def __init__(self, name, age):
        self.name = name  # Attribute
        self.age = age    # Attribute
    
    def bark(self):      # Method
        return "Woof!"

my_dog = Dog("Buddy", 3)  # creates an object of the Dog class
print(my_dog.name)
print(my_dog.bark())
```
Why review classes and functions? Because some things in Decisions Trees persist while others do not. Be sure to consider which elements/capabilities are more appropriately represented as a class vs. a function.
